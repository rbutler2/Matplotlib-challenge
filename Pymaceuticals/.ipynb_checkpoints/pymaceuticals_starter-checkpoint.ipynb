## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_mouse_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")
# Display the data table for preview
merge_mouse_df
```

```python
# Checking the number of mice.
num_o_mice = len(merge_mouse_df["Mouse ID"].unique())
num_o_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
#unique_mouse = merge_mouse_df.drop_duplicates(subset = "Mouse ID", keep = "first", inplace= False)
#unique_mouse

duplicate_mice = []
dup_mice = merge_mouse_df.duplicated(subset= ["Mouse ID", "Timepoint"])
dups = merge_mouse_df[dup_mice]
dups_m_id = dups["Mouse ID"].unique()

duplicate_mice.insert(0, dups_m_id)
print(f"duplicate mice by ID: {duplicate_mice}")
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

clean_mouse = merge_mouse_df.drop(merge_mouse_df[merge_mouse_df["Mouse ID"] == "g989"].index)
clean_mouse
clean_mouse_df = pd.DataFrame(clean_mouse)
clean_mouse_df
```

```python
# Checking the number of mice in the clean DataFrame.
num_clean_mice = len(clean_mouse_df["Mouse ID"].unique())
print(f"The number of mice in clean DataFrame: {num_clean_mice}")
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

summ_stat_mice = clean_mouse_df.groupby("Drug Regimen")["Tumor Volume (mm3)"]
mean_drug_tumor = summ_stat_mice.mean()
median_drug_tumor = summ_stat_mice.median()
var_drug_tumor = summ_stat_mice.var()
std_drug_tumor = summ_stat_mice.std()
sem_drug_tumor = summ_stat_mice.sem()

summ_stat_df = pd.DataFrame({"Mean": mean_drug_tumor, "Median": median_drug_tumor, "Standard Deviation": std_drug_tumor, "SEM": sem_drug_tumor})
summ_stat_df
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
capo_df = clean_mouse_df[["Mouse ID","Tumor Volume (mm3)", "Timepoint"]]
desired_mouse = ("x401",)
rcapo_df = capo_df[capo_df["Mouse ID"].isin(desired_mouse)]
lines = rcapo_df.plot.line()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
wvt_df = clean_mouse_df[["Mouse ID", "Drug Regimen", "Tumor Volume (mm3)", "Weight (g)"]]
desired_drug = ("Capomulin",)
only_capo = wvt_df[wvt_df["Drug Regimen"].isin(desired_drug)]
only_capo
only_capo_at = only_capo.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()
only_capo_at_df = pd.DataFrame(only_capo_at)
mouse_weight = only_capo.groupby("Mouse ID")["Weight (g)"].max()
mouse_weight_df = pd.DataFrame(mouse_weight)
#only_capo_wvt = pd.DataFrame({"Average Tumor Volume (mm3)": [only_capo_at], "Mouse Weight (g)": [mouse_weight]})
only_capo_wvt = pd.merge(only_capo_at_df, mouse_weight_df, on= "Mouse ID")
only_capo_wvt_plot = only_capo_wvt.plot.scatter(x="Tumor Volume (mm3)", y="Weight (g)", c="b")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
variable = st.pearsonr(only_capo_wvt["Tumor Volume (mm3)"], only_capo_wvt["Weight (g)"])
corr = variable[0]
print(f"The correlation coefficient between Tumor Volume and Average Weight is {corr}")
x = only_capo_wvt.iloc[:,0]
y = only_capo_wvt.iloc[:,1]

wvt_slope, wvt_int, wvt_r, wvt_p, wvt_str_eer = st.linregress(x,y)

wvt_fit = wvt_slope * x + wvt_int

x = only_capo_wvt.iloc[:,0].values.reshape(-1,1)
y = only_capo_wvt.iloc[:,1].values.reshape(-1,1)
plt.scatter(x,y)
plt.plot(x,wvt_fit,"--")
plt.xlabel("Tumor Volume (mm3)")
plt.ylabel("Weight(g)")
plt.show()
```
